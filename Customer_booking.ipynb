```python

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt



# Convert categorical columns to categorical types (if not done already)
categorical_columns = ['sales_channel', 'trip_type', 'route', 'booking_origin']
df[categorical_columns] = df[categorical_columns].astype('category')

# One-hot encode categorical variables
df_encoded = pd.get_dummies(df, columns=categorical_columns, drop_first=True)

# Define features and target
X = df_encoded.drop('booking_complete', axis=1)  # Features
y = df_encoded['booking_complete']  # Target variable

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt

# Assume X_train is your feature set and y_train is your target variable

# Step 1: Train a Random Forest model to compute feature importances
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Step 2: Extract and rank feature importances
importances = model.feature_importances_
feature_names = X_train.columns

# Create a DataFrame for feature importances
importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances})
importance_df = importance_df.sort_values(by='Importance', ascending=False)

# Print the feature importances
print("Feature Importances:")
print(importance_df)

# Step 3: Select Top Features
# Define the number of top features to select
n_top_features = 10
top_features = importance_df.head(n_top_features)['Feature']

# Step 4: Transform the dataset to use only the top features
X_train_selected = X_train[top_features]

# Optionally, visualize the feature importances
plt.figure(figsize=(12, 8))
plt.barh(importance_df['Feature'], importance_df['Importance'], color='skyblue')
plt.xlabel('Importance')
plt.title('Feature Importance')
plt.gca().invert_yaxis()  # To display the most important features at the top
plt.show()

print(f"Selected Top {n_top_features} Features:")
print(top_features)
```
