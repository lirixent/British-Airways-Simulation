```python
import requests
from bs4 import BeautifulSoup
import csv

# Base URL of the page containing the reviews (replace with actual base URL)
base_url = 'https://www.airlinequality.com/airline-reviews/british-airways'

# Define the number of pages to scrape
total_pages = 383

# List to hold review data
reviews_data = []

# Iterate through each page
for page in range(1, total_pages + 1):
    # Construct the URL for the current page
    if page == 1:
        url = base_url
    else:
        url = f'{base_url}/page/{page}/'

    # Send a GET request to fetch the page content
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find all review sections on the current page
    review_sections = soup.find_all('div', class_='text_content')

    # Iterate over each review section to extract details
    for review_section in review_sections:
        # Extract review body
        review_body = review_section.get_text(strip=True)

        # Find the ratings and recommendation within the same review section
        review_stats = review_section.find_next_sibling('div', class_='review-stats')
        if review_stats:
            rating_table = review_stats.find('table', class_='review-ratings')
            if rating_table:
                ratings = {}
                rating_elements = rating_table.find_all('tr')

                for element in rating_elements:
                    header = element.find('td', class_='review-rating-header')
                    stars = element.find('td', class_='review-rating-stars')
                    value = element.find('td', class_='review-value')

                    if header:
                        header_text = header.get_text(strip=True)

                        if stars:
                            # Count the number of filled stars
                            star_elements = stars.find_all('span', class_='star')
                            rating = sum(1 for star in star_elements if 'fill' in star.get('class', []))
                            ratings[header_text] = f"{rating} out of 5"
                        elif value:
                            # Handle non-star ratings
                            value_text = value.get_text(strip=True)
                            ratings[header_text] = value_text

                # Extract recommendation
                recommended = review_stats.find('td', class_='review-value rating-no')
                recommended_text = recommended.get_text(strip=True) if recommended else 'N/A'

                # Append extracted data to the list
                reviews_data.append({
                    'Review Body': review_body,
                    'Seat Comfort': ratings.get('Seat Comfort', 'N/A'),
                    'Cabin Staff Service': ratings.get('Cabin Staff Service', 'N/A'),
                    'Food & Beverages': ratings.get('Food & Beverages', 'N/A'),
                    'Ground Service': ratings.get('Ground Service', 'N/A'),
                    'Value For Money': ratings.get('Value For Money', 'N/A'),
                    'Recommended': recommended_text
                })

# Write the data to a CSV file
csv_file = 'airline_reviews.csv'
csv_columns = ['Review Body', 'Seat Comfort', 'Cabin Staff Service', 'Food & Beverages', 'Ground Service', 'Value For Money', 'Recommended']

with open(csv_file, 'w', newline='', encoding='utf-8') as file:
    writer = csv.DictWriter(file, fieldnames=csv_columns)
    writer.writeheader()
    writer.writerows(reviews_data)

print(f"Data has been exported to {csv_file}")
```
